```python
plt.figure(figsize=(7,6))
sns.regplot(x='stress level', y='snoring rate', data=pillow)
plt.xticks([0,1,2,3,4],['low/normal', 'medium low', 'medium', 'medium high', 'high'])
plt.ylabel('Snoring Rate')
plt.xlabel('Stress Level')
plt.title('Effect of Snoring Rate on Stress Level')
```
